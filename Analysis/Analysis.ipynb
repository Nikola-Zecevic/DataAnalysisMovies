# Exploring Factors Influencing Movie Success!

First, let's load cleaned data!

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../Data/movies_cleaned.csv')

df.head()
```

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

### Hypothesis 1: Bigger budget leads to bigger gross!

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='budget', y='gross', alpha=0.5)
plt.title('Budget vs Gross')
plt.xlabel('Budget')
plt.ylabel('Gross')
plt.grid(True)
plt.savefig('../Images/budget_vs_gross.png')
plt.show()
```

Let us check for correlation between the two!

```python
correlation = df['budget'].corr(df['gross'])
print(f"Correlation between budget and gross is:{correlation:.2f}")
```

Conclusion for Hypothesis 1:
Plot shows positive connection between budget and gross. 
Correlation is high(0.74) which indicates that movies with high budget have high gross.
Hypothesis 1 is confirmed!

![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

### Hypothesis 2: Movies with higher score have higher gross!

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='score', y='gross', alpha=0.5)
plt.title('Score vs Gross')
plt.xlabel('Score')
plt.ylabel('Gross')
plt.grid(True)
plt.savefig('../Images/score_vs_gross.png')
plt.show()
```

Let's check for correlation between these two!

```python
correlation_score_gross = df['score'].corr(df['gross'])
print(f"Correlation between score and gross is:{correlation_score_gross:.2f}")
```

Conclusion for Hypothesis 2:
Based on plot, there isn't strong visual connection between score and gross!
Correlation is 0.2 which means that it is low positive correlation between score and gross!
Hypothesis 2 is partially true, but score is not strong predictor of high grossing movie!


![purple-divider](https://user-images.githubusercontent.com/7065401/52071927-c1cd7100-2562-11e9-908a-dde91ba14e59.png)

### Hypothesis 3: Certain genres are more profitable!

```python
genre_gross = df.groupby('genre')['gross'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=genre_gross.values, y=genre_gross.index)
plt.title('Average gross per genre')
plt.xlabel('Average gross')
plt.ylabel('Genre')
plt.grid(True)
plt.savefig('../Images/gross_by_genre.png')
plt.show()
```

Conclusion for Hypothesis 3:
Visuelisation of average gross per genre shows that certain genres achieve significantly higher average gross!
Therefore, hipothesis 3 is confirmed: genre has influence on gross!
